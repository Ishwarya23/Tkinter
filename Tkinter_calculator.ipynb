```python
from tkinter import *
calwindow = Tk()
calwindow.geometry("350x250")
calwindow.title("CALCULATOR")
calwindow.resizable(0,0)

inputvar = StringVar()
expression = ""

def btnclicked(item):    
    global expression
    expression = expression + str(item)
    inputvar.set(expression)
    
def cleartext():
    global expression
    expression = ""
    inputvar.set(expression)
    
def evalresult():
    global expression
    result = str(eval(expression)) 
    inputvar.set(result)
    expression = ""
    
inputframe = Frame(calwindow,width=350,height=20,highlightbackground="black")
inputframe.pack(side = TOP)
inputtext = Entry(inputframe,justify = RIGHT,font =('arial', 18, 'bold'),textvariable = inputvar)

inputtext.grid(row = 0, column = 0)
inputtext.pack(ipady = 10)

buttonframe = Frame(calwindow, width = 350, height = 260, bg = "lightgrey")
buttonframe.pack()

#Row1
clrbt = Button(buttonframe,text="Clear",width=37,height=2, bd = 0, bg = "#eee", cursor = "hand2",command=lambda: cleartext()).grid(row = 0, column = 0, columnspan = 3,padx = 1, pady = 1)
divbt = Button(buttonframe,text = "/", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked("/")).grid(row = 0, column = 3, padx = 1, pady = 1)

#Row2
sevenbt = Button(buttonframe,text = "7", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(7)).grid(row = 1, column = 0, padx = 1, pady = 1)
eightbt= Button(buttonframe,text = "8", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(8)).grid(row = 1, column = 1, padx = 1, pady = 1)
ninebt= Button(buttonframe,text = "9", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(9)).grid(row = 1, column = 2, padx = 1, pady = 1)
multiplybt= Button(buttonframe,text = "*", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked("*")).grid(row = 1, column = 3, padx = 1, pady = 1)

#Row3
fourbt = Button(buttonframe,text = "4", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(4)).grid(row = 2, column = 0, padx = 1, pady = 1)
fivebt= Button(buttonframe,text = "5", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(5)).grid(row = 2, column = 1, padx = 1, pady = 1)
sixbt= Button(buttonframe,text = "6", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(6)).grid(row = 2, column = 2, padx = 1, pady = 1)
subbt= Button(buttonframe,text = "-", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked("-")).grid(row = 2, column = 3, padx = 1, pady = 1)

#Row4
onebt = Button(buttonframe,text = "1", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(1)).grid(row = 3, column = 0, padx = 1, pady = 1)
twobt= Button(buttonframe,text = "2", width=12,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(2)).grid(row = 3, column = 1, padx = 1, pady = 1)
threebt= Button(buttonframe,text = "3", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(3)).grid(row = 3, column = 2, padx = 1, pady = 1)
addbt= Button(buttonframe,text = "+", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked("+")).grid(row = 3, column = 3, padx = 1, pady = 1)

#Row5
zerobt= Button(buttonframe,text = "0", width=25,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(0)).grid(row = 4, column = 0, columnspan = 2, padx = 1, pady = 1)
decibt= Button(buttonframe,text = ".", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: btnclicked(".")).grid(row = 4, column = 2, padx = 1, pady = 1)
equalbt= Button(buttonframe,text = "=", width=11,height=2,bd = 0, bg = "#eee", cursor = "hand2",command=lambda: evalresult()).grid(row = 4, column = 3, padx = 1, pady = 1)

calwindow.mainloop()
```
